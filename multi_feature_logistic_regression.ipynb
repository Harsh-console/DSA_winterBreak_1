```python
import numpy as np
```

```python
import pandas as pd
data = pd.read_csv('./assests/logistic_regression_data.csv')
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
y.shape
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def logistic_regression(X, y):
    m, n = X.shape
    # m, 1 = y.shape

    # initialize parameters
    w = np.random.rand(n, 1)
    b = np.random.rand()
    
    #hyper paramters
    learning_rate = 0.01
    n_iters = 1000
    for i in range(n_iters):
        z = X.dot(w) + b
        y_hat = sigmoid(z)
        temp = y_hat - y
        grad_w = (1 / m) * X.T.dot(temp)
        grad_b = (1 / m) * np.sum(temp)
        w -= learning_rate *grad_w
        b -= learning_rate *grad_b
    return w, b
```

```python
w, b = logistic_regression(X, y)
```
